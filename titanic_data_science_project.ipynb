##### **2.2 Explain the method(s) you used to address it (e.g., dropping, imputing)**

**1. Age (177 missing values)**
- Method: Imputing using the median age grouped by Title, Sex, and Pclass
- Reasoning: Age varies significantly by title (ex. Miss, Ms, Mrs, Master, Mr) and Pclass

```python
#Extract the titles from the "Name" column
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)

df['Age'] = df.groupby(['Title', 'Pclass', 'Sex'])['Age'].transform(lambda x: x.fillna(x.median()))

print(df['Age'].isnull().sum())
```

**2. Embarked (2 missing values)**  
- Method: Fill using the mode (most common port)  
- Reasoning: Only 2 values are missing, so using the mode is simple and does not distort the data significantly.

```python
df['Embarked'].fillna(df['Embarked'].mode()[0])
```

**3. Cabin (687 missing values)**

- Reasoning: Too much missing values result in unreliable analysis or not significant results

### **3. Transformation - Feature Engineering & Variable Exclusion**

##### **3.1 Describe the logic behind any new features and how they helped**


```python
# Extract last name from the Name column
df['LastName'] = df['Name'].str.extract(r'([A-Za-z]+),', expand=False)

# Calculate family size based on SibSp and Parch (immediate family on board)
df['FamilySize_SibParch'] = df['SibSp'] + df['Parch'] + 1

# Count how many times each last name appears (extended family/group)
last_name_counts = df['LastName'].value_counts()
df['FamilySize_LastName'] = df['LastName'].map(last_name_counts)

# Choose the maximum of the two as the family size feature
df['FamilySize'] = df[['FamilySize_SibParch', 'FamilySize_LastName']].max(axis=1)

print(df[['FamilySize']].value_counts().sort_index())
```

```python
#Make Age column is all float type
df['Age'] = df['Age'].astype(float)

#Define bins and labels for age groups
age_bins = [0.0, 1.0, 15.0, 25.0, 45.0, 65.0, 100.0]
age_labels = ['Infant', 'Children', 'Youth', 'Adult', 'MiddleAge', 'Senior']

#Create a new column 'AgeGroup' based on the defined bins and labels
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)

#Count the occurrences of each age group
print(df[['AgeGroup']].value_counts().sort_index())

#Count total number of sum of each age group
print(df['AgeGroup'].value_counts().sum())
```

### **4. Explatory Analysis - Survival Patterns**

##### **4.1 Identify and describe patterns in survival based on passenger characteristics**

```python
survived_by_age = df.groupby('Age')["Survived"].mean()

# Show survival rate by each age with percentage in index
trend = survived_by_age.apply(lambda x: f"{x:.2%}")
trend.index = [f"Age {age} ({pct})" for age, pct in zip(trend.index, trend.values)]
print(trend.sort_index())
```

**1. Survival by Age**

The analysis shows infants nearly had 100% of survival rate. This could be assumed due to evacuation priorities goes to women and children. General trend of Survival by Age is shown in following graph. Higher survival rates goes to younger ages and older ages which showing trends of bimodal shape. 

```python
survived_by_sex = df.groupby('Sex')['Survived'].mean()

# Show survival rate by gender with percentage in index
trend = survived_by_sex.apply(lambda x: f"{x:.2%}")
trend.index = [f"{gender} ({pct})" for gender, pct in zip(trend.index, trend.values)]
print(trend)
```

**2. Survival by Gender**

The analysis shows a big difference in survival rates between men and women. Women had a much higher chance of survival (74%) compared to men (19%). This matches historical accounts that women were given priority during the evacuation of the Titanic. Gender was the single most important factor in the dataset.

```python
survived_by_Pclass = df.groupby('Pclass')['Survived'].mean()

#Show survival rate by Pclass with percentage in index
trend = survived_by_Pclass.apply(lambda x: f"{x:.2%}")
trend.index = [f"Class {cls}" for cls in trend.index]
print(trend)
```

**3. Survival by Passenger Class (Pclass)**

Survival rates varied by passenger class. First class had the highest survival rate, second class was second and third class had the lowest. This matches the social hierarchy on board where first class had better cabin access, more lifeboats nearby and possibly faster access to evacuation information.

```python
survived_by_familysize = df.groupby('FamilySize')['Survived'].mean()

# Show survival rate by FamilySize with size and percentage in index
trend = survived_by_familysize.apply(lambda x: f"{x:.2%}")
trend.index = [f"Family Size {size}" for size in trend.index]
print(trend)
```

**4. Survival by Family Size**

Family size had a non-linear impact on survival. Solo passengers (family size = 1) had lower survival rates possibly because they had no one to help them during the chaos. Those with family sizes of 2 to 4 had the highest survival chances — they were together and could help each other. Very large families (5+) had lower survival rates possibly because of coordination issues or being in cabins further from the lifeboats.

```python
survived_by_agegroup = df.groupby('AgeGroup')['Survived'].mean()

# Show survival rate by AgeGroup with group and percentage in index
trend = survived_by_agegroup.apply(lambda x: f"{x:.2%}")
trend.index = [f"{group} ({pct})" for group, pct in zip(trend.index, trend.values)]
print(trend)
```

**5. Survival by Age Group**

When binned into age groups, infants (under 1 year old) had the highest survival rate, children next and teenagers and adults had much lower survival rates. This supports the narrative that children were also prioritized during the rescue. The elderly had the lowest survival rates possibly due to mobility issues and slower response times.
